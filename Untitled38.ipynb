```python
class SSMLNode:
  def __init__(self,text="",children=None):
    self.text = text
    if children is None:
      children = []
    self.children = children

  def ssml_node_to_text(root):
    result=[]

    def dfs(node):
      if node.text:
        result.append(node.text)

        for child in node.children:
          dfs(child)

    dfs(root)
    return "".join(result)
```
